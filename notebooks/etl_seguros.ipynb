bloque 3 limpieza de datos

```python


def limpiar_dataframe(df):

    df.columns = df.columns.str.strip().str.lower()

    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.strip()

    df = df.replace(r'^\s*$', pd.NA, regex=True)

    df = df.drop_duplicates()

    return df
```

```python
#FUNCIONES GENERALES DE LIMPIEZA
# --------------------------------------------

def normalizar_columnas(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
    )
    return df


def limpiar_texto_basico(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace(
            ["nan", "None", "NULL", "null", ""],
            pd.NA
        )
    return df


def eliminar_duplicados(df):
    return df.drop_duplicates()


def convertir_fecha_multiple(serie):
    return pd.to_datetime(serie, errors="coerce", dayfirst=False)


def limpiar_numero_mixto(valor):

    if pd.isna(valor):
        return np.nan

    valor = str(valor).strip()

    if valor in ["", "-", "None", "nan", "NULL", "null"]:
        return np.nan

    if "," in valor and "." in valor:

        if valor.rfind(",") > valor.rfind("."):
            valor = valor.replace(".", "")
            valor = valor.replace(",", ".")
        else:
            valor = valor.replace(",", "")

    elif "," in valor:
        valor = valor.replace(",", ".")

    try:
        return float(valor)
    except:
        return np.nan
```

```python
#NORMALIZACIÓN INICIAL
# --------------------------------------------

dfs = [clientes, aseguradoras, corredores, tipos_seguro, polizas, siniestros]

for i in range(len(dfs)):
    dfs[i] = normalizar_columnas(dfs[i])
    dfs[i] = limpiar_texto_basico(dfs[i])
    dfs[i] = eliminar_duplicados(dfs[i])

clientes, aseguradoras, corredores, tipos_seguro, polizas, siniestros = dfs
```

```python
#LIMPIEZA DE CLIENTES
# --------------------------------------------

clientes["email"] = clientes["email"].str.lower()

map_genero = {
    "hombre": "Masculino",
    "masculino": "Masculino",
    "m": "Masculino",
    "mujer": "Femenino",
    "femenino": "Femenino",
    "f": "Femenino"
}

clientes["genero"] = clientes["genero"].str.lower().map(map_genero)


map_ciudad = {
    "sanmiguel": "San Miguel",
    "s. miguel": "San Miguel",
    "san miguel": "San Miguel",
    "sta. ana": "Santa Ana",
    "sta ana": "Santa Ana",
    "santa ana": "Santa Ana",
    "ss": "San Salvador",
    "san salvador": "San Salvador"
}

clientes["ciudad"] = clientes["ciudad"].str.lower().replace(map_ciudad)
clientes["ciudad"] = clientes["ciudad"].str.title()

clientes["fecha_nacimiento"] = convertir_fecha_multiple(clientes["fecha_nacimiento"])

clientes["segmento"] = clientes["segmento"].fillna("No especificado")


clientes_validos = clientes[
    clientes["id_cliente"].notna() &
    clientes["nombre"].notna() &
    clientes["email"].notna()
].copy()

clientes_rechazados = clientes[
    clientes["id_cliente"].isna() |
    clientes["nombre"].isna() |
    clientes["email"].isna()
].copy()


```

```python
## 5. LIMPIEZA DE ASEGURADORAS
# --------------------------------------------

aseguradoras["nombre"] = aseguradoras["nombre"].str.title()
aseguradoras["pais"] = aseguradoras["pais"].str.title()

map_rating = {
    "alto": "Alto",
    "medio": "Medio",
    "bajo": "Bajo"
}

aseguradoras["rating_riesgo"] = aseguradoras["rating_riesgo"].str.lower().map(map_rating)
aseguradoras["rating_riesgo"] = aseguradoras["rating_riesgo"].fillna("No especificado")
aseguradoras["pais"] = aseguradoras["pais"].fillna("No especificado")
```

```python
# 6. LIMPIEZA DE CORREDORES
# --------------------------------------------

corredores["nombre"] = corredores["nombre"].str.title()
corredores["zona"] = corredores["zona"].str.strip().str.title()

map_nivel = {
    "junior": "Junior",
    "mid": "Mid",
    "senior": "Senior"
}

corredores["nivel"] = corredores["nivel"].str.strip().str.lower().map(map_nivel)
corredores["nivel"] = corredores["nivel"].fillna("No especificado")

corredores["anios_experiencia"] = pd.to_numeric(
    corredores["anios_experiencia"], errors="coerce"
)
```

```python
# 7. LIMPIEZA DE TIPOS DE SEGURO
# --------------------------------------------

tipos_seguro["tipo"] = tipos_seguro["tipo"].str.title()
tipos_seguro["categoria"] = tipos_seguro["categoria"].str.title()
tipos_seguro["categoria"] = tipos_seguro["categoria"].fillna("No especificado")

tipos_seguro["riesgo_base"] = tipos_seguro["riesgo_base"].apply(limpiar_numero_mixto)


```

```python
# 8. LIMPIEZA DE POLIZAS
# --------------------------------------------

polizas["fecha_emision"] = convertir_fecha_multiple(polizas["fecha_emision"])

polizas["prima"] = polizas["prima"].apply(limpiar_numero_mixto)
polizas["comision"] = polizas["comision"].apply(limpiar_numero_mixto)
polizas["monto_asegurado"] = polizas["monto_asegurado"].apply(limpiar_numero_mixto)

```

```python
# 9. LIMPIEZA DE SINIESTROS
# --------------------------------------------

siniestros["fecha_siniestro"] = convertir_fecha_multiple(siniestros["fecha_siniestro"])
siniestros["monto_siniestro"] = siniestros["monto_siniestro"].apply(limpiar_numero_mixto)

map_estado = {
    "abierto": "Abierto",
    "cerrado": "Cerrado",
    "en proceso": "En Proceso",
    "pendiente": "Pendiente"
}

siniestros["estado"] = siniestros["estado"].str.strip().str.lower().map(map_estado)
siniestros["estado"] = siniestros["estado"].fillna("No especificado")
```
